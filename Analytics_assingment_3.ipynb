```python
#Libraries used in this exercise

import numpy as np
import pandas as pd
```

```python
#EXCERSISE 1

#1.1 Import the data and create a dataframe which consists the columns of interest:
movie_data = pd.read_csv('rotten_tomatoes_top_movies.csv')
columns = ['title', 'year', 'critic_score', 'people_score', 'total_reviews', 'total_ratings', 'rating', 'type', 'original_language', 'director', 'release_date_(theaters)', 'runtime']

topmovie_data = pd.DataFrame(movie_data[columns])
topmovie_data.head()
```

```python
#1.2 
#What is the average number of reviews for the movies included in the dataset (total reviews column)? 
#After calculating this value, filter your data for movies that have at least as many reviews as this average value.
```

```python
#The average number of reviews for the movies is 143.65217391304347
average_num_reviews = topmovie_data['total_reviews'].mean()

#Below is a dataframe that consists of movies that have at least as many reviews as the average value.
topmovie_data[topmovie_data.total_reviews >= average_num_reviews]
```

```python
#1.3
#How many rows do you have in this data in which the movie is more than 3 hours long? 
#(Hint: what is the first character/digit of the runtime column if the movie is more than 3 hours long?
```

```python
# First, I will filter out all the rows in which the string '3h' appears in the column 'runtime'.
duration = topmovie_data['runtime'].str.contains('3h', case = False, na = False)

# Then, from this amount of rows I will reduce the amount of movies(rows) that are exactly 3h long, since
# we are only interested about movies that are more than 3h long. The answer is 27 rows'
topmovie_data[duration].shape[0] - topmovie_data[topmovie_data.runtime == '3h'].shape[0]
```

```python
#1.4
#Choose your favourite type, and filter your data to include only movies of that type. 
#Perform the following tasks on this filtered data.
#For which movie in this dataset you can find the largest difference between the critics and viewers evaluation (critic score and people score)?
```

```python
#First, I will see which specific movie types there are in the dataset by using the .unique() function.
topmovie_data['type'].unique()

#Then I select my favourite movie type out of the dataset. 
my_type = topmovie_data[topmovie_data['type'] == 'Action & Adventure'] 

#Then I select all the movie titles out of the chosen dataset. 
#By using function .unique() I make sure there are no duplicates.

titles = my_type.title.unique()

#Then I calculate the difference between the critic_score and people_score values for the specific movies. 
comp = my_type.critic_score -  my_type.people_score

#Then I create a dictionary with the movie titles and critic & people score differences for dataframe purposes. 
data = {'Movie title':titles,'critics_viewers_evaluation_diff':comp}

#Then I pass in the 'data' variable as an argument into the DataFrame function and create a dataframe.  
my_dataframe = pd.DataFrame(data)

#Then I sort the dataframe by the absolute difference values and choose the first value row in the dataframe.
my_dataframe.sort_values(by = 'critics_viewers_evaluation_diff', ascending = False).head(1)

#The answer in my case is 'Star Wars: The Last Jedi', it has the largest difference in critic and people evaluation.


```

```python
#How many directors do you have in the dataset with more than one movie?

# First, I will count how many times different directors appear in the dataset. 
directors = topmovie_data[topmovie_data['type'] == 'Action & Adventure'].director.value_counts()

#Then, I count the times when the directors appears more than once in the selected dataset. 
(directors >1).value_counts()

#In my dataset there are 19 directors with more than one movie. 
```

```python
#2.4
#Check the correlation between revenue and number of employees. 
#Based on the results, what is your opinion, does the size of the company in terms of employees impact revenue? 
#If you calculate the correlation between profit and number of employees, would you make the same observation?
```

```python
corr_one = fortune_data['revenue'].corr(fortune_data['num. of employees'])
corr_two = fortune_data['profit'].corr(fortune_data['num. of employees'])

print("Correlation between revenue and num. of employees is ", corr_one)
print("Correlation between profit and num. of employees is ", corr_two)

# Based on the results, I would say that the number of employees within a company 
# plays a significant role in revenue generation since the result is closer to one and so
# the corelation between the two variables is strong.
# However when it comes to the profit I would say that it has less significance since the
# result is closer to zero and thus weaker correlation.
```

```python
#2.5
# How many companies do you find on the list that made losses, 
# even though based on revenue they are in the top 1000? 
# What is the biggest loss (i.e., most negative profit) that you find in the data, 
# and in what sector is the company that made that loss?
```

```python
# Sorting the dataset based on revenue.
top_thousand = fortune_data.sort_values(by = ["revenue"], ascending = False)


# After sorting, we can select all the companies that made negative profit.
# There are 114 companies with negative profit within the 1000 best companies by revenue.
top_thousand[top_thousand.profit < 0].shape[0]

# We can see that the biggest loss is (-6520.0) and the sector is 'Industrials'
biggest_loss = top_thousand[top_thousand.profit < 0].sort_values(by = ['profit'])[:1]
biggest_loss
```

```python
# 2.6
# Create a categorical version of the revenue column with 4 groups: (i) values between 0 and 3500, 
# (ii) values between 3500 and 6380, (iii) values between 6380 and 14620, and (iv) values between 14620 and 1000000.
# Label the groups [’Low’, ’Medium’, ’High’, ’Very High’]. In which category do you find the most companies?
# Which sector has the highest average revenue?
```

```python
# Creating the categorical version of the revenue column
bins = [0,3500,6380,14620,1000000]

groups = ['Low','Medium','High','Very High']

fortune_data['revenue_group'] = pd.cut(fortune_data.revenue,4, bins, labels=groups)

# most of the companies belong to the 'low' category
fortune_data['revenue_group'].value_counts().sort_values(ascending = False)[:1]

# The 'very high' category has the highest average revenue
low = fortune_data[fortune_data.revenue_group == "Low"].revenue.mean()
med = fortune_data[fortune_data.revenue_group == "Medium"].revenue.mean()
high = fortune_data[fortune_data.revenue_group == "High"].revenue.mean()
veryhigh = fortune_data[fortune_data.revenue_group == "Very High"].revenue.mean()

print(low,med,high,veryhigh)
```
